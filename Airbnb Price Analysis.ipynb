# Exploratory Data Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data Preprocessing

```python
ams_weekdays = pd.read_csv('amsterdam_weekdays.csv').dropna()
ams_weekends = pd.read_csv('amsterdam_weekends.csv').dropna()
amsterdam = pd.concat([ams_weekdays, ams_weekdays])

barcelona_weekdays = pd.read_csv('barcelona_weekdays.csv').dropna()
barcelona_weekends = pd.read_csv('barcelona_weekends.csv').dropna()
barcelona = pd.concat([barcelona_weekdays, barcelona_weekends])

athens_weekdays = pd.read_csv('athens_weekdays.csv').dropna()
athens_weekends = pd.read_csv('athens_weekends.csv').dropna()
athens = pd.concat([athens_weekdays, athens_weekends])

berlin_weekdays = pd.read_csv('berlin_weekdays.csv').dropna()
berlin_weekends = pd.read_csv('berlin_weekends.csv').dropna()
berlin = pd.concat([berlin_weekdays, berlin_weekends])

budapest_weekdays = pd.read_csv('budapest_weekdays.csv').dropna()
budapest_weekends = pd.read_csv('budapest_weekends.csv').dropna()
budapest = pd.concat([budapest_weekdays, budapest_weekends])

lisbon_weekdays = pd.read_csv('lisbon_weekdays.csv').dropna()
lisbon_weekends = pd.read_csv('lisbon_weekends.csv').dropna()
lisbon = pd.concat([lisbon_weekdays, lisbon_weekends])

london_weekdays = pd.read_csv('london_weekdays.csv').dropna()
london_weekends = pd.read_csv('london_weekends.csv').dropna()
london = pd.concat([london_weekdays, london_weekends])

paris_weekdays = pd.read_csv('paris_weekdays.csv').dropna()
paris_weekends = pd.read_csv('paris_weekends.csv').dropna()
paris = pd.concat([paris_weekdays, paris_weekends])

rome_weekdays = pd.read_csv('rome_weekdays.csv').dropna()
rome_weekends = pd.read_csv('rome_weekends.csv').dropna()
rome = pd.concat([rome_weekdays, rome_weekends])

vienna_weekdays = pd.read_csv('vienna_weekdays.csv').dropna()
vienna_weekends = pd.read_csv('vienna_weekends.csv').dropna()
vienna = pd.concat([vienna_weekdays, vienna_weekends])
```

## Average Price Listing

```python
data = {'City': ['Amsterdam', 'Barcelona', 'Athens', 'Berlin', 'Budapest', 'Lisbon', 'London', 'Paris', 'Rome', 'Vienna'],
                  'Average Price': [amsterdam['realSum'].mean(),
barcelona['realSum'].mean(),
athens['realSum'].mean(),
berlin['realSum'].mean(),
budapest['realSum'].mean(),
lisbon['realSum'].mean(),
london['realSum'].mean(),
paris['realSum'].mean(),
rome['realSum'].mean(),
vienna['realSum'].mean()]}

average_prices = pd.DataFrame(data).sort_values(by='Average Price', ascending=False)
print(average_prices)

plt.barh(average_prices['City'], average_prices['Average Price'], edgecolor='k', color='darkorange')
plt.xlabel('Average Price (in Euros)')
plt.title('Rank Average Price Listings In 10 European Countries')
plt.show()
```

# Price Distribution in Athens

```python
athens_weekdays = pd.read_csv('athens_weekdays.csv').dropna()
athens_weekends = pd.read_csv('athens_weekends.csv').dropna()
athens = pd.concat([athens_weekdays, athens_weekends])
pd.DataFrame(athens['realSum'].describe())

max_value = athens['realSum'].max()
athens = athens[athens['realSum'] != max_value]
pd.DataFrame(athens['realSum'].describe())

mean = athens['realSum'].mean()
median = athens['realSum'].median()
mode = athens['realSum'].mode()[0]
plt.hist(athens['realSum'], edgecolor='k', color='mediumseagreen')
plt.axvline(mean, linestyle='dashed', color='red', linewidth=2, label=f'Average Price: {mean:.2f}')
plt.axvline(median, linestyle='dashed', color='k', linewidth=2, label=f'Typical Price: {median:.2f}')
plt.axvline(mode, linestyle='dashed', color='blue', linewidth=2, label=f'Most Common Price: {mode:.2f}')
plt.xlabel('Price Listings (in Euros)')
plt.ylabel('Counts')
plt.title('Distribution of Renting Prices in Athens')
plt.legend()
plt.show()
```

# Room Types

```python
europe_cities = pd.concat([athens, budapest, rome, lisbon, berlin, vienna, barcelona, london, paris, amsterdam])
rooms = europe_cities['room_type'].value_counts()
labels = europe_cities['room_type'].unique()
plt.pie(rooms, autopct='%1.1f%%',
       wedgeprops={"edgecolor":"k",'linewidth': 1,'antialiased': True})
plt.tight_layout()
plt.legend(labels, bbox_to_anchor=(0.9, 1), loc='upper left', borderaxespad=0)
plt.show()
```

```python
home_apt = europe_cities[europe_cities['room_type'] == 'Entire home/apt']
private_room = europe_cities[europe_cities['room_type'] == 'Private room']
shared_room = europe_cities[europe_cities['room_type'] == 'Shared room']
print('Average price listing of Entire Home/Apartment:', home_apt['realSum'].mean())
print('Average price listing of Private Rooms:', private_room['realSum'].mean())
print('Average price listing of Shared Rooms:', shared_room['realSum'].mean())
```
